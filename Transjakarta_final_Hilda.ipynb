#### **Mengisi `corridorID` berdasarkan `corridorName`**

```python
#mapping utk liat identitas koridor berdasarkan nama dan id koridor
missing_corridor_id_mapping = df.dropna(subset=['corridorID']).drop_duplicates('corridorName')
missing_corridor_id_mapping = missing_corridor_id_mapping.set_index('corridorName')['corridorID']
missing_corridor_id_mapping.reset_index()
```

```python
#isi corridor id - corridor name function
def fill_corridor_id(row):
    if pd.isnull(row['corridorID']) and pd.notnull(row['corridorName']):
        return missing_corridor_id_mapping.get(row['corridorName'])
    return row['corridorID']

#apply fill function
df['corridorID'] = df.apply(fill_corridor_id, axis=1)
```

#### **Mengisi missing values `corridorID` dengan `tapInStopsLat` dan `tapInStopsLon`**

```python
#mapping utk liat identitas koridor berdasarkan nama dan id koridor
corridor_id_latlong = df.dropna(subset=['corridorID']).drop_duplicates(['tapInStopsLat', 'tapInStopsLon'])
corridor_id_latlong = corridor_id_latlong.set_index(['tapInStopsLat', 'tapInStopsLon'])['corridorID']
corridor_id_latlong.reset_index()
```

```python
#isi corridor id - latlong function
def fill_missing_corridor_id(row):
    if pd.isnull(row['corridorID']) and pd.notnull(row['tapInStopsLat']) and pd.notnull(row['tapInStopsLon']):
        return corridor_id_latlong.get((row['tapInStopsLat'], row['tapInStopsLon']))
    return row['corridorID']

#apply function
df['corridorID'] = df.apply(fill_missing_corridor_id, axis=1)
```

#### **Mengisi missing values `corridorName` dengan `corridorID`**

```python
#mapping corridor name dengan corridor id
corridor_name_mapping = df.dropna(subset=['corridorName']).drop_duplicates('corridorID')
corridor_name_mapping = corridor_name_mapping.set_index('corridorID')['corridorName']
corridor_name_mapping.reset_index()
```

```python
#isi corridor name dengan corridor id function
def fill_missing_corridor_name(row):
    if pd.isnull(row['corridorName']) and pd.notnull(row['corridorID']):
        return corridor_name_mapping.get(row['corridorID'])
    return row['corridorName']

#apply function
df['corridorName'] = df.apply(fill_missing_corridor_name, axis=1)
```

#### **Mengisi missing values `payAmount` dan menambah kolom baru `transType`**
selanjutnya kita akan mengisi missing values `payAmount` berdasarkan tipe kendaraan.

Tipe kendaraan Transjakarta akan dibagi 3 berdasarkan `payAmount` dan diberi nama `transType` dengan penjelasan berikut:
- Busway = IDR 3500 atau biaya general/flat yang dibayar pengguna Transjakarta
- RoyalTrans = IDR 20000, hanya lewat untuk beberapa rute 
- Mikrotrans = IDR 0

```python
#buat kolom baru untuk tipe kendaraan nya
def tipe_trans (payAmount) :
    if payAmount == 3500:
        return 'Busway'
    elif payAmount == 20000:
        return 'RoyalTrans'
    else:
        return 'Mikrotrans'
    
df['transType'] = df['payAmount'].apply(tipe_trans)

df.head()
```

```python
#Mengecek penumpang yang memakai Royal Trans (bayar IDR 20000) ada di koridor mana saja
df.loc[df['payAmount'] == 20000.0]['corridorName'].unique()
```

```python
corridor_royal = ['Cibubur - Balai Kota', 'Bekasi Barat - Kuningan',
       'Palem Semi - Bundaran Senayan', 'Cinere - Bundaran Senayan',
       'Bintaro - Fatmawati', 'Senen - Blok M',
       'Stasiun Tebet - Karet via Patra Kuningan', 'Cibubur - BKN',
       'Bekasi Barat - Blok M', 'Cinere - Kuningan',
       'Cibubur Junction - Blok M', 'Cibubur - Kuningan',
       'BSD Serpong - Fatmawati', 'Tanah Abang - Blok M']

#fungsi fill pay amount
def fill_pay_amount(row):
    if row['corridorName'] in corridor_royal and pd.isna(row['payAmount']):
        return 20000
    elif pd.isna(row['payAmount']):
        return 3500
    else:
        return row['payAmount']
    
#apply function
df['payAmount'] = df.apply(fill_pay_amount, axis=1)
```

### Koridor teramai di morning rush hour

```python
#buat kolom baru utk rush hour
def rush_hour (tapInHour) :
    if 5 <= tapInHour <= 9:
        return 'Morning Rush Hour'
    elif 16 <= tapInHour <= 20:
        return 'Evening Rush Hour'
    else:
        return 'Non Rush Hour'
    
df['rush_hour'] = df['tapInHour'].apply(rush_hour)

#cek 
display (df.sample(3))
```

```python
#morning rush hour & evening rush hour

# jam 5 - 9
morning_rush_hour = df[df['rush_hour'] == 'Morning Rush Hour'].groupby(['corridorID', 'corridorName']).size().reset_index(name='count')
morning_rush_hour = morning_rush_hour.sort_values(by='count', ascending=False).head()
display (morning_rush_hour) 

#-----------------------------------------------------------------------------------------------------------------------------------------#

#jam 16 - 20
evening_rush_hour = df[df['rush_hour'] == 'Evening Rush Hour'].groupby(['corridorID', 'corridorName']).size().reset_index(name='count')
evening_rush_hour = evening_rush_hour.sort_values(by= 'count', ascending=False).head()
display (evening_rush_hour)
```

```python
#plotting koridor based on morning rush hour & evening rush hour

# Morning Rush Hour Plot
plt.figure(figsize=(20, 5))
plt.bar(morning_rush_hour['corridorName'], morning_rush_hour['count'], color='#6eadf0')
plt.title('Morning Rush Hour - Top Corridors')
plt.xlabel('Nama Koridor')
plt.ylabel('Jumlah Pengguna')
plt.show()

#--------------------------------------------------------------------------------------------------------------------------------------#

# Evening Rush Hour Plot
plt.figure(figsize=(20, 5))
plt.bar(evening_rush_hour['corridorName'], evening_rush_hour['count'], color='#f0af6e')
plt.title('Evening Rush Hour - Top Corridors')
plt.xlabel('Nama Koridor')
plt.ylabel('Jumlah Pengguna')
plt.show()
```
